```python
import torch
```

Useful functions

```python
# choose random integer in the range [0, n)
choose_from = lambda n: torch.randint(n, (1,))[0]

def create_map(num_people,num_fam,choose_from):
    r"""Array mapping people to families
    size: (num_people, num_fam)
    """
    out = torch.zeros((num_people,num_fam))

    for p in range(num_people):
        n = choose_from(num_fam)
        out[p, n] = 1

    return out

def ground_truth(people_fam_map: torch.Tensor) -> float:
    r"""Returns the average family size.
    dim 0: # of people
    dim 1: # of families
    """
    assert people_fam_map.dim() == 2, "input must be a 2D array"

    fam_sizes = people_fam_map.sum(0)

    out = fam_sizes.sum() / len(fam_sizes)
    return float(out)
```

Sampling strategies

```python
NUM_PEOPLE = 10000
NUM_FAM = 2500

# mapping array
people_fam_map = create_map(NUM_PEOPLE, NUM_FAM, choose_from)

"""1. SRATEGY FROM QUESTION"""

total = 0
NUM_SAMPLES = 370
for _ in range(NUM_SAMPLES):
    # 1. Select a person uniformly at random, say i
    i = choose_from(NUM_PEOPLE)

    # 2. Note down the size of the family the person i belongs to, let's call it x_i
    fam_id = list(people_fam_map[i]).index(1) # family id
    x_i = people_fam_map[:, fam_id].sum()

    total += x_i
# 3. Take the average of x_i for i = 1,2,3,...n, let's call it x_mean
x_mean = total/NUM_SAMPLES


"""2. PROPOSED STRATEGY"""
NUM_SAMPLES = 333
total = 0
for _ in range(NUM_SAMPLES):
    # 1. Select a family uniformly at random, say i
    i = choose_from(NUM_FAM)

    # 2. Note down the size of that family
    fam_size = people_fam_map[:, i].sum()

    total += fam_size
# 3. Take the average
avg_prop = total/NUM_SAMPLES

"""3. GROUND TRUTH"""
avg_truth = ground_truth(people_fam_map)



print('The average size of a family:')
print(f'\t Strategy from question: {x_mean:.2f}')
print(f'\t Proposed strategy: {avg_prop:.2f}')
print(f'\t Ground truth: {avg_truth}')
```
